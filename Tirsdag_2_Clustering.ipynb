```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage 
from sklearn import metrics
from sklearn.metrics import davies_bouldin_score
```

```python
# Load data and make a simple linkage with Wards linkage
df=pd.read_csv("https://github.com/CALDISS-AAU/sdsphd19_coursematerials/raw/master/data/data.csv")
X = df[['criminals', 'diffrace', 'leftwings', 'alcoholics', 'rightwings', 'manykids', 'neurotics', 'muslims', 'imigrants', 'aids', 'addicts', 'homoseksuals', 'jews', 'gypsies']]
X_1 = df[['criminals', 'diffrace', 'leftwings', 'alcoholics', 'rightwings', 'manykids', 'neurotics', 'muslims', 'imigrants', 'aids', 'addicts', 'homoseksuals', 'jews', 'gypsies']]
linked = linkage(X, 'ward')
print(X)
```

```python
# Look at data and make a dendogram 
figure = plt.figure(figsize=(7.5, 5))
dendrogram(
    linked,
    truncate_mode='lastp',  # show only the last p merged clusters
    p=20,  # show only the last p merged clusters
    leaf_rotation=90.,
    leaf_font_size=12.,
    show_contracted=True,  # to get a distribution impression in truncated branches
)
plt.title('Hierarchical Clustering Dendrogram (Ward)')
plt.xlabel('sample index or (cluster size)')
plt.ylabel('distance')
plt.show()
```

```python
for k in range(2,11):
    cluster = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage='ward').fit(X) # Gentager med 2-11 (k) clusters med agglomerative
    labels=cluster.labels_
    dbs=davies_bouldin_score(X, labels) # fremgår i outputtet db, jo lavere jo bedre (1 er godt)
    dbs=round(dbs,2)
    ch = metrics.calinski_harabasz_score(X, labels) # ch-score, jo større jo bedre
    ch=round(ch,2)
    ss=metrics.silhouette_score(X, labels, metric='euclidean') # 1 er bedst, -1 værst. Måler minimeret varians indeni klyngerne og maksimeret mellem klyngerne
    ss=round(ss,2)
    print("Cluster count-", k, "|", " db score-", dbs, "|", " ch score - ", ch, "|", " ss score - ", ss)
    k=str(k) #* Ændrer k til en tekststreng for at kunne sætte den på variablen nedenfor
    df['cluster'+k]=labels
    
    #* Outputtet peger på 2,3 eller 10 clusters
```

```python
for i in range(2,6):
	i=str(i) # i til en streng for at kunne anvendes i navngivning
	X_1=df[['cluster'+i,'criminals', 'diffrace', 'leftwings', 'alcoholics', 'rightwings', 'manykids', 'neurotics', 'muslims', 'imigrants', 'aids', 'addicts', 'homoseksuals', 'jews', 'gypsies', 'age', 'income', 'edu']]
	stuff=X_1.groupby(['cluster'+i]).mean() # Gennemsnit af alle variable indenfor clusteret
	trans_stuff=stuff.T # Transformeret så tabellen vender rigtigt
	ts=round(trans_stuff,2)
	print("")
	print("")
	print('... and the results for' ,i, 'clusters are!')	
	print(ts)
	print("")
	count=df.groupby(['cluster'+i])['id'].count()
	print("And the size of the clusters if there are",i, "clusters are")
	print(count)
	print("")
	print("And if we have",i,"clusters, the dendrogram looks like")
	i=int(i)
	figure = plt.figure(figsize=(7.5, 5))
	dendrogram(
	        linked,
		truncate_mode='lastp',  # show only the last p merged clusters
		p=i,  # show only the last p merged clusters
		leaf_rotation=90.,
		leaf_font_size=12.,
		show_contracted=True,  # to get a distribution impression in truncated branches
	)
	plt.title('Hierarchical Clustering Dendrogram (Ward)')
	plt.xlabel('sample index or (cluster size)')
	plt.ylabel('distance')
	plt.show()
```
